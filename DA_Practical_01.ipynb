```python
from matplotlib import pyplot as plt
```

```python
plt.plot(df['Name'],df['SE'])
plt.title("linear graph",fontsize=25,color="green")
plt.xlabel('Student name')
plt.ylabel('SE')
plt.ylim(0,100)
plt.legend(["marks in SE"])
plt.show()
```

```python
plt.bar(df['Name'],df['SE'])
plt.title("bar graph",fontsize=25,color="green")
plt.xlabel('Student name')
plt.ylabel('SE')
plt.legend(["marks in SE"])
plt.show()
```

```python
plt.bar(df['Name'],df['SE'])
plt.bar(df['Name'],df['OOPS'])
plt.title("bar graph",fontsize=25,color="green")
plt.xlabel('Student name')
plt.show()
```

```python
plt.scatter(df['Name'],df['SE'],c="blue")
plt.scatter(df['Name'],df['OOPS'],c="red")
plt.scatter(df['Name'],df['HCI'],c="green")
plt.scatter(df['Name'],df['CNW'],c="black")
plt.scatter(df['Name'],df['ES'],c="orange")
plt.show()
```
